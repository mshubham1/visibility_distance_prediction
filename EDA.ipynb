
# **Data Preprocessing**

```python
#distribution of data
for column in df.drop(['DATE'],axis=1):
  sns.distplot(df[column])
  plt.figure(figsize=(10,10))
```

```python
df.corr()
```

# Feature Selection

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
```

```python
# Here the feature StationPressure and SeaLevelPressure are similar, 
from sklearn.ensemble import ExtraTreesRegressor
feature_selection = ExtraTreesRegressor()
feature_selection.fit(df.drop(['DATE','VISIBILITY'],axis=1),df['VISIBILITY'])
```

```python
print(feature_selection.feature_importances_)
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(feature_selection.feature_importances_,index =df.drop(['DATE','VISIBILITY'],axis=1).columns)
feat_importances.nlargest(14).plot(kind='barh')
plt.show()
```

# Droping the feature which are highly corelated

```python
X=df.drop(['DATE','VISIBILITY','DRYBULBTEMPF','DewPointTempF','StationPressure'],axis=1)
y=df['VISIBILITY']
print(X.columns)
```

```python
for feature in X.columns:
    sns.boxplot(X[feature])
    plt.title(feature)
    plt.figure(figsize=(10,7))
```

```python
# Drop the Precip as it has all values range to 0
X_new=X.drop(['Precip'],axis=1)
```

```python
X_new.columns
```

```python
X_new.head()
```

```python
#Distribution of data
for feature in X_new.columns:
    sns.distplot(X_new[feature])
    plt.title(feature)
    plt.figure(figsize=(10,7))
```

# Model Building

```python
##Model Building
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2,random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
# DecisionTreeRegressor
decision_model=DecisionTreeRegressor(max_depth=20)
decision_model.fit(X_train,y_train)
decision_model.score(X_train,y_train)
```

```python
decision_model.score(X_test,y_test)
```

```python
y_pred=decision_model.predict(X_test)
```

```python
from sklearn import metrics
metrics.r2_score(y_test, y_pred)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Random Forest Regressor

```python
# RandomForestRegressor
randomforest_model=RandomForestRegressor(max_depth=25)
randomforest_model.fit(X_train,y_train)
randomforest_model.score(X_train,y_train)
```

```python
randomforest_model.score(X_test,y_test)
```

```python
y_pred=randomforest_model.predict(X_test)
```

```python
from sklearn import metrics
metrics.r2_score(y_test, y_pred)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# XGB Regressor

```python
##XGBRegressor
xbg_model=XGBRegressor(n_estimators=100,max_depth=20)
xbg_model.fit(X_train,y_train)
xbg_model.score(X_train,y_train)
```

```python
xbg_model.score(X_test,y_test)
```

```python
y_pred=xbg_model.predict(X_test)
```

```python
from sklearn import metrics
metrics.r2_score(y_test, y_pred)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
